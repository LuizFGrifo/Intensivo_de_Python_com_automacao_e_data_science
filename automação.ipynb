```python
import pyautogui
import time

# Acessa o painel do Windows
pyautogui.press('win')
# Procura pelo navegador
pyautogui.write('edge')
# E entra no navegador
pyautogui.press('Enter')

# Pause entre as ações da biblioteca
pyautogui.PAUSE = 1

# Atalho para uma nova aba
pyautogui.hotkey('ctrl', 't')
# Passo 1: Acessar o sistema da empresa
time.sleep(2)
pyautogui.write('https://pages.hashtagtreinamentos.com/aula1-intensivao-sistema')
# Para entrar no link
pyautogui.press('Enter')

# Para esperar o site carregar
time.sleep(5)

# Para clickar na caixa de login
pyautogui.click(x=807, y=579)
# Para escrever o login
pyautogui.write('meu login')
# Para clickar na caixa de senha
pyautogui.click(x=794, y=686)
# Para escrever a senha
pyautogui.write('minha senha')
# Passo2: para clickar no botão "Acessar"
pyautogui.click(x=937, y=804)

# Para esperar o site carregar
time.sleep(3)

# Passo 3: baixar o arquivo csv
pyautogui.click(x=741, y=564, button='right') # button do botão esquerdo
time.sleep(2)
pyautogui.click(x=941, y=851)

# Caso demore para baixar a base de dados
time.sleep(3) 
```

```python
# Passo 4: Calcular os indicadores
import pandas as pd

# Importar a base de dados
tabela = pd.read_csv(r"C:\Users\lfgri\OneDrive\Documentos\Programação\Python\intensivão de Python\Arquivos\Compras.csv", sep=';')
display(tabela) # É um print mais bonito. ps: só tem no jupyter
# Caculo dos indicadores
# Total gasto -> somar a coluna quantidade
total_gasto = tabela["ValorFinal"].sum()
# Quantidae -> somar a coluna quantidade
quantidade = tabela["Quantidade"].sum()
# Preço médio
preco_medio = total_gasto / quantidade

print(total_gasto)
print(quantidade)
print(preco_medio)
```

```python
# Passo 5: Enviar o email para o diretorio/para o chefe
import pyautogui
import pyperclip
import time

pyautogui.hotkey('ctrl', 't')
pyautogui.click(x=1325, y=110)
pyautogui.click(x=1348, y=232)
time.sleep(1)
pyautogui.click(x=3015, y=308)
time.sleep(1)
pyautogui.write('luizgrifo465@hotmail.com')
time.sleep(1)
pyautogui.press('Enter')
pyautogui.press('tab')
pyperclip.copy('Enviando o relatório da aula 1')
pyautogui.hotkey('ctrl', 'v')
time.sleep(1)
pyautogui.press('tab')
texto = f"""
Prezado Luiz Felipe Grifo,

segue o relatorio de compras

Total Gasto: R${total_gasto:,.2f}
Quantidade de Produtos: {quantidade:,} 
Preço Médio: R${preco_medio:,.2f}

Qualquer dúvida estarei à disposição!
Att. Escravo da Grifo's Corp
"""
pyperclip.copy(texto)
pyautogui.hotkey('ctrl', 'v')

time.sleep(1)
pyautogui.hotkey('ctrl', 'Enter') # Para enviar o email
```
